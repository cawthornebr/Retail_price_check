```python
# set up browser
url = "https://www.homedepot.com/b/Appliances/N-5yc1vZbv1w"

browser = webdriver.Chrome()
```

```python
# go to url and navigate to first set of refrigerators
browser.get(url)

#select refrigerators from the appliance section
appliances = browser.find_elements_by_class_name("visualNav__title-center")

appliances[0].click()

#select french door refrigerators
time.sleep(0.08)
fridge = browser.find_elements_by_class_name("etchVisualNavigationStory__card__title")
fridge[0].click()
```
